```python
import sqlite3
import requests
import xmltodict 
```

```python

response = requests.get('https://knesset.gov.il/Odata/ParliamentInfo.svc/KNS_BillUnion()') #   Based on the question was asked in the group by Neta
dict_data = xmltodict.parse(response.content)
# print(type(dict_data))
# print(dict_data)
dict_data
```

```python
db = sqlite3.connect('my_database.db')
cursor = db.cursor()
cursor.execute(
        '''
        CREATE TABLE IF NOT EXISTS KNS_BillName(
        BillUnionID INTEGER PRIMARY KEY,
        MainBillID INTEGER, 
        UnionBillID INTEGER ,  
        LastUpdatedDate TEXT)
        ''' )
db.commit()
atributes =[]
```

```python
for element in dict_data['feed']['entry']:
    atribute = element['content']['m:properties']
    BillUnionID = int(atribute['d:BillUnionID']['#text'])
    MainBillID= int(atribute['d:MainBillID']['#text'])
    UnionBillID = int(atribute['d:UnionBillID']['#text'])
    LastUpdatedDate = atribute['d:LastUpdatedDate']['#text']
    atributes.append(( MainBillID,UnionBillID, LastUpdatedDate))
```

```python
# atributes = [BillNameID, Name,NameHistoryTypeID, NameHistoryTypeDesc, LastUpdatedDate]
print(len(atributes))
atributes[:100]
```

```python
cursor.executemany('''
                INSERT INTO KNS_BillName(

MainBillID,
UnionBillID,
LastUpdatedDate
)
                 VALUES (?,?,?)''', atributes)
cursor.execute('''
                   SELECT * FROM KNS_BillName
                   ''')
```
